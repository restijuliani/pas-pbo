```python
import matplotlib.pyplot as plt # VISUALISASI DATA

df.plot(kind='scatter', x='Life expectancy', y='percentage expenditure') # UNTUK NAMA VARIABLE 
plt.title('Life expectancy')
plt.xlabel('percentage expenditure')
plt.ylabel('indeks')
plt.xlim(0, 69.1)
plt.ylim(0,17)
plt.grid(True)
plt.show()
```

```python
import numpy as np
x = np.array (df['Life expectancy'])
y = np.array (df['percentage expenditure'])

print(f'x : {x}')
print(f'y : {y}')
```

```python
from sklearn.linear_model import LinearRegression

model =LinearRegression()
model.fit(x, y)

```
